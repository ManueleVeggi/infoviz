```python
from SPARQLWrapper import SPARQLWrapper, JSON
import sparql_dataframe
import pandas as pd

endpoint = "http://data.fondazionezeri.unibo.it/sparql"

query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX crm: <http://www.cidoc-crm.org/cidoc-crm/>
PREFIX ti: <http://www.ontologydesignpatterns.org/cp/owl/timeinterval.owl#>
SELECT DISTINCT ?esemplare ?label_artwork ?trasferimento ?label_receiver 
WHERE {
  <https://w3id.org/zericatalog/the-metropolitan-museum-of-art-new-york-ny> crm:P29i_received_custody_through ?trasferimento .
  ?trasferimento crm:P30_transferred_custody_of ?esemplare .
  ?trasferimento crm:P28_custody_surrendered_by ?surrender .
  ?receiver crm:P29i_received_custody_through ?trasferimento . 
  
  ?esemplare  rdfs:label ?label_artwork.
  ?trasferimento rdfs:label ?label_transfer.
  ?surrender rdfs:label ?label_surrender.
  ?receiver rdfs:label ?label_receiver.
  
  FILTER(LANG(?label_artwork) = "" || LANGMATCHES(LANG(?label_artwork), "en"))
  FILTER(LANG(?label_transfer) = "" || LANGMATCHES(LANG(?label_transfer), "en"))
  FILTER(LANG(?label_surrender) = "" || LANGMATCHES(LANG(?label_surrender), "en"))
  FILTER(LANG(?label_receiver) = "" || LANGMATCHES(LANG(?label_receiver), "en"))

}
"""

df = sparql_dataframe.get(endpoint, query, post=True)
df
```

```python
a_string = ""
for idx, row in df.iterrows():
    if idx == 0:
            a_string = "?esemplare = <" + row["esemplare"] + ">"
    else: a_string = a_string + "||" + "?esemplare = <" + row["esemplare"] + ">"

new_query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX crm: <http://www.cidoc-crm.org/cidoc-crm/>
PREFIX ti: <http://www.ontologydesignpatterns.org/cp/owl/timeinterval.owl#>
SELECT ?esemplare ?trasferimento_global ?receiver ?label_receiver ?surrender ?label_surrender
WHERE {{
  ?trasferimento_global crm:P30_transferred_custody_of ?esemplare .
  ?trasferimento_global crm:P28_custody_surrendered_by ?surrender .
  ?receiver crm:P29i_received_custody_through ?trasferimento_global .
  FILTER ({0})
  
  ?esemplare  rdfs:label ?label_artwork.
  ?trasferimento_global rdfs:label ?label_transfer.
  ?surrender rdfs:label ?label_surrender.
  ?receiver rdfs:label ?label_receiver.
  
  FILTER(LANG(?label_artwork) = "" || LANGMATCHES(LANG(?label_artwork), "en"))
  FILTER(LANG(?label_transfer) = "" || LANGMATCHES(LANG(?label_transfer), "en"))
  FILTER(LANG(?label_surrender) = "" || LANGMATCHES(LANG(?label_surrender), "en"))
  FILTER(LANG(?label_receiver) = "" || LANGMATCHES(LANG(?label_receiver), "en"))
}}
"""
new_df = sparql_dataframe.get(endpoint, new_query.format(a_string), post=True)
new_df
```

```python
new_string = ""
for idx, row in df.iterrows():
    if idx == 0:
        new_string = new_string + "strstarts(str(?custody), '" + row["esemplare"].replace("/item", "/custody") + "')"
    else:
        new_string = new_string + "|| strstarts(str(?custody), '" + row["esemplare"].replace("/item", "/custody") + "')"

access = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX crm: <http://www.cidoc-crm.org/cidoc-crm/>
PREFIX ti: <http://www.ontologydesignpatterns.org/cp/owl/timeinterval.owl#>
SELECT ?custody ?time_span ?end_date
WHERE {{
  ?receiver crm:P29i_received_custody_through ?custody .
  ?custody crm:P4_has_time_span ?time_span .
  ?time_span ti:hasIntervalEndDate ?end_date .
  FILTER ({0})
}}
"""
third_df = sparql_dataframe.get(endpoint, access.format(new_string), post=True)
third_df
```

```python
import re

year_dict = {}
for idx, row in third_df.iterrows():
    number = re.findall("\d+\d", row["custody"])
    num = number[0]
    if num not in year_dict: 
        year_dict[num] = row["end_date"][:4]
    else:
        if int(year_dict[num]) < int(row["end_date"][:4]):
            year_dict[num] == row["end_date"][:4]

year_list = []
for key, value in year_dict.items():
    year_list.append(value)
```

```python
URI_string = ""

for key, value in year_dict.items():
    URI = "https://w3id.org/zericatalog/artwork/" + key + "/item"

    if len(URI_string) == 0:
        URI_string = "?esemplare = <" + URI + ">"
    else: 
        URI_string = URI_string + "||" + "?esemplare = <" + URI + ">"

final_query = """
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX crm: <http://www.cidoc-crm.org/cidoc-crm/>
PREFIX ti: <http://www.ontologydesignpatterns.org/cp/owl/timeinterval.owl#>
SELECT DISTINCT ?esemplare ?label_artwork 
WHERE {{
  ?receiver crm:P29i_received_custody_through ?trasferimento .
  ?trasferimento crm:P30_transferred_custody_of ?esemplare .
  FILTER({0})
  
  ?esemplare  rdfs:label ?label_artwork.
  
  FILTER(LANG(?label_artwork) = "" || LANGMATCHES(LANG(?label_artwork), "en"))
}}

"""
final_df = sparql_dataframe.get(endpoint, final_query.format(URI_string), post=True)
final_df['year'] = year_list

final_df
```

```python
artists = {}


for idx, row in final_df.iterrows():
    row["label_artwork"] = row["label_artwork"].replace("Artwork", "").replace("- exemplar", "").replace('"', "")
    row["label_artwork"] = row["label_artwork"].strip().split (" , ")
    if len(row["label_artwork"][0].split(" ")) == 2:
        a_list = [row["label_artwork"][0].split(" ")[1] + " " + row["label_artwork"][0].split(" ")[0]]
        a_list.append(row["label_artwork"][1])
        row["label_artwork"] = a_list
    if row["label_artwork"][0] not in artists:
        artists[row["label_artwork"][0]] = {row["esemplare"]: row["year"]}
    else:
        artists[row["label_artwork"][0]][row["esemplare"]] = row["year"]

final_df
artists



```
